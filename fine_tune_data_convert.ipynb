```python
import json

def load_jsonl(path):
    data = []
    with open(path, "r", encoding="utf-8") as f:
        for line in f:
            line = line.strip()
            if line:
                data.append(json.loads(line))
    return data
```

```python
def to_chat_format(example):
    return {
        "messages": [
            {
                "role": "user",
                "content": example["input"]
            },
            {
                "role": "assistant",
                "content": json.dumps(example["output"], ensure_ascii=False)
            }
        ]
    }
```

```python
train_dataset = load_jsonl("train_dataset.jsonl")
test_dataset = load_jsonl("test_dataset.jsonl")

train_chat_dataset = [to_chat_format(x) for x in train_dataset]
test_chat_dataset = [to_chat_format(x) for x in test_dataset]
```

```python
# save ready to train data into files
def save_jsonl(data, path):
    with open(path, "w", encoding="utf-8") as f:
        for item in data:
            f.write(json.dumps(item, ensure_ascii=False) + "\n")

save_jsonl(train_chat_dataset, "train_chat.jsonl")
save_jsonl(test_chat_dataset, "test_chat.jsonl")
```
